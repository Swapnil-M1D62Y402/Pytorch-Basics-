##Reshaping, stacking, squeezing, unsqueezing tensors

1. Reshaping - reshape an input tensor to a defined shape
2. View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
3. Stacking - combine mutiple tensors on top of each other(vstack) or side by side
(hstack)
4. Squeeze - removes all `1` dimensions from a tensor
5. Unsqueeze - add a `1` dimension to a target tensor
6. Permute - Return a view of the input with dimensiosn permuted in a certain way

```python
x = torch.arange(1., 10.)
print(x)
print(x.shape)
```

```python
x_reshape = x.reshape(3,3)  # 3*3 =shape([9])
x_reshape
```

```python
x_reshaped = x.reshape(1,9) # 1*9 = 9
print(x_reshaped)
print(x_reshaped.shape)
```

```python
x_reshaped = x.reshape(9,1) #transposes the matrix # 9*1 = 9
print(x_reshaped)
print(x_reshaped.shape)
```

```python
#change the view
z = x.view(1,9)
print(z)
print(z.shape)
```

**Changing z changes x (because a view of a tensor shares the same memory as the original tensor)**

Its like reference in C++

```python
z[:,0] = 5
print("z = ", z)
print("x = ", x)
```

```python
#torch.permute - rearrange the dimensions
x_original = torch.rand(size=(224,224,3)) #[height,width,color_channel]

#permute the orignal tensor to rearrange the axis (or dim ) order
x_permuted = x_original.permute(2,0,1,)

print(f"Original tensor {x_original}and shape{x_original.shape}")
print(f"Permuted tensor {x_permuted}and shape{x_permuted.shape}")
```

```python
x_original[0,0,0] = 0.69
x_permuted = x_original.permute(2,0,1,)
print(x_original[0,0,0,])
print(x_permuted[0,0,0,])
#hence we see that the change is copied to the x_permuted
#thus permuted uses a view of the original and any changed done to the original
#is reflected to the copy also
```

##Reproducbility (trying to take random out of random tensors)

In short how a neural network learns:

`start with random numbers -> tensor operations performed -> update the random numbers to fit the model -> Repeat the steps`

However in a reproducable model, you dont want much randomness, which could harm the reproducable nature of the model. Thats when we use `**random.seed()**`
Essentially what random seed does is "flavour" the randomness

```python
import torch
rand_tensor_A = torch.rand(3,3)
rand_tensor_B = torch.rand(3,3)

print(f"First Tensor : {rand_tensor_A}")
print(f"Second Tensor :  {rand_tensor_B}")

print(f"Are first and second tensors equal? {rand_tensor_A == rand_tensor_B}")
```

```python
#Set random seed
RANDOM_SEED = 1
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3,4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3,4)

print(f"First Tensor : {random_tensor_C}")
print(f"Second Tensor :  {random_tensor_D}")
print(f"Are first and second tensors equal? {random_tensor_C == random_tensor_D}")
```
